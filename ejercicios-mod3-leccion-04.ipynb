2. Los nombres de las columnas no son homogeneos. Cambia los nombres de las columnas de tal forma que:

```python
"""No tengan espacios."""
"""Estén en minúscula."""
"""No tengan paréntesis, es decir, quitar "(%)", "(Km2)"."""
"""Algunas columnas tiene "\n". Eliminalos de los nombres de las columnas."""
"""Algunas columnas tienen ":". Eliminalos de los nombres de las columnas."""

nuevo_df = {columna: columna.lower().replace(" ", "_").replace("(%)", "%").replace("\n", "").replace(":", "") for columna in mergeado_izq.columns}

nuevo_df
```

```python
mergeado_izq.rename(columns = nuevo_df, inplace = True)

# comprobamos que se han cambiado los nombres de las columnas
mergeado_izq.head()
```

```python
mergeado_izq.rename(columns={"density(p/km2)":"density_p/km2","land_area(km2)":"land_area_km2", "agricultural_land(_%)":"agricultural_land_%"}, inplace = True)

# comprobamos que se han cambiado los nombres de las columnas
mergeado_izq.head()
```

```python
"""La columnas coordinates tiene la latitud y la longitud en una sola columna. Crea dos columnas nuevas, una con la longitud y otra con la latitud. Una vez hecho, elimina la columna de coordinates."""

mergeado_izq["coordinates"].str.split(",")
```

```python
mergeado_izq["coordinates"].str.split(",", expand = True)
```

```python
mergeado_izq[["latitud", "longitud"]] = mergeado_izq["coordinates"].str.split(",", expand = True).get([0,1])

# comprobamos que se han creado correctamente las columnas
mergeado_izq.head(2)
```

```python
mergeado_izq.drop("coordinates", axis=1, inplace=True)
```

```python
mergeado_izq.head(2)
```

```python
"""Las columnas unemployment_rate, total_tax_rate, tax_revenue, population_labor_force_participation, out_of_pocket_health_expenditure, gross_tertiary_education_enrollment, gross_primary_education_enrollment, forested_area, cpi_change, agricultural_land tienen "%". Elimina los "%" de los valores de las columnas."""

mergeado_izq["unemployment_rate"] = mergeado_izq["unemployment_rate"].str.replace("%", " ")

mergeado_izq.head()
```

```python
columnas_a_modificar = ["total_tax_rate", "tax_revenue_%", "population_labor_force_participation_%", "out_of_pocket_health_expenditure", "gross_tertiary_education_enrollment_%", "gross_primary_education_enrollment_%", "forested_area_%", "cpi_change_%", "agricultural_land_%"]  # Agrega aquí todas las columnas que necesites

for columna in columnas_a_modificar:
    if columna in mergeado_izq.columns:
        mergeado_izq[columna] = mergeado_izq[columna].str.replace("%", " ")

mergeado_izq.head()
   
```

```python
"""Haz lo mismo para las columnas de gasoline_price, gdp, minimum_wage, pero eliminando "$"."""

mergeado_izq["gasoline_price"] = mergeado_izq["gasoline_price"].str.replace("$", " ")

# chequeamos que el cambio se hizo correctamente
mergeado_izq.head()
```

```python
columnas_a_modificar = ["gdp", "minimum_wage"]  # Agrega aquí todas las columnas que necesites

for columna in columnas_a_modificar:
    if columna in mergeado_izq.columns:
        mergeado_izq[columna] = mergeado_izq[columna].str.replace("$", " ")

mergeado_izq.head()
```

```python
mergeado_izq["latitud"] = mergeado_izq["latitud"].str.replace("(", " ")

mergeado_izq.head()
```

```python
mergeado_izq["longitud"] = mergeado_izq["longitud"].str.replace(")", " ")

mergeado_izq.head()
```

```python
"""Guarda el DataFrame para usarlo posteriormente."""

mergeado_izq.to_csv("df_world_limpieza.csv", index=False)
```
